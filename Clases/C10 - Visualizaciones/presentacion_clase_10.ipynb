### Figures y axes

### Figures y axes

---
Observemos como creamos primero una *Figure* (<code>plt.figure</code>), luego un *Axe* (<code>plt.axes</code>) y finalmente un gráfico de líneas sobre el axe (<code>ax.plot</code>).

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
ax = plt.axes()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

```python
%matplotlib inline
fig, axs = plt.subplots(3, 3, figsize=(6, 2))  # a figure with a 2x2 grid of Axes
# figsize: tamaño para todos los axes.
```

### Figures y axes. Subplot

---
El método `plt.subplot()` permite crear simultáneamente la figura y los axes que lo componen:

<a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html" target="blank">https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html</a>

**Nota**: para visualizar los gráficos creados con matplotlib tenemos dos opciones:

* `%matplotlib inline` muestra los gráficos **debajo** de la celda de código ejecutada. Se ejecuta **sólo una vez** por kernel / sesión.

* `plt.show()` muestra **el último** gráfico creado, independientemente de la ubicación de la celda.

### Grillas. Grid

---
El método `plt.grid()` permite crear grillas sobre los axes.

<a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.grid.html" target="blank">https://matplotlib.org/api/_as_gen/matplotlib.pyplot.grid.html</a>

Observar que se pueden referenciar cada uno de los axes.

- Axe [0,0] líneas continuas (es el default), solo sobre el eje x, azul.
- Axe [0,1] linea de trazos, sobre ambos ejes (es el default), rojo y ancho 2.
- Axe [1,0] línea punteada, solo sobre el eje y, verde.

```python
fig, axs = plt.subplots(2,2, figsize=(8, 2), sharex=True, sharey=True)
# sharex, sharey: comparten todos los axes las escalas de x e y
axs[0,0].grid(axis='x', color='blue')
axs[0,1].grid(color='r', linestyle='--', linewidth=2)
axs[1,0].grid(axis='y', color='g', linestyle=':')
plt.show()
```

### Etiquetas

---
Matplotlib tiene métodos que permiten generar distintas etiquetas del gráfico.
 
* Título de la figura:  `Figure.suptitle`

* Título de los ejes y del plot: `Axes.set` con `xlabel`, `ylabel`, `title`

```python
fig = plt.figure(figsize=(5, 4))
ax = plt.axes()
fig.suptitle("título de la figura")
ax.set(xlabel = "título eje x ", ylabel = "título eje y", title = "titulo del plot")
plt.show()
```

### Leyendas

---
Al crear el gráfico de líneas con `plot`, con el parámetro `label` indicamos que nombre tiene cada serie. Y generamos una leyenda que indique en el gráfico esos nombres con `legend`.

Podemos indicar el lugar donde colocamos la leyenda (`loc`) y el título (`title`).

<a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html" target="blank">https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html</a>

```python
import numpy.random as rd
rd.seed(0)
random1 = rd.randn(30).cumsum() # Creamos un array de 30 números según la distribución normal 
random2 = rd.randn(30).cumsum()

fig = plt.figure(figsize=(6, 2))
ax = plt.axes()

ax.plot([20, 10, 5], color = '0.75', label='random1')
#ax.plot(random2, color = 'g',  label='random2')
ax.legend(loc='upper right', title = "series")
plt.show()
```

### Ajuste de ejes

---
Se pueden indicar los intervalos en los cuales queremos graficar los ejes x e y.

Se usan los métodos `set_xlim` y `set_ylim`.

<a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlim.html" target="blank">https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlim.html</a>


```python
# El mismo ejemplo anterior, limitando el eje x e y
fig = plt.figure(figsize=(6, 2))
ax = plt.axes()

ax.set_xlim(5, 12)    
ax.set_ylim(-5, 10)
#ax.set(xlim=(5, 12), ylim= (-5, 10)) # Otra forma
ax.plot(random1, color = '0.75', label='random1')
ax.plot(random2, color = 'g',  label='random2')
plt.show()
```

### Grafico de líneas (Plot)
---
Ya estuvimos viendo como crear este gráfico, mediante el método `plot`. Agreguemos a la definición del color y el tipo de linea, como destacar los *puntos de la serie o marcas* con `marker`. 

Observar como definimos el color y la línea de otra forma: `k--`.

Tambien se pueden modificar los puntos del eje x con `set_xticks` (los valores a mostrar) y con que labels `set_xticklabels`. Similarmente se puede hacer con el eje y.

Para ver como llamar a los colores, tipos de líneas y marcas: <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html" target="blank">https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html</a>

```python
rd.seed(10)
random1 = rd.randn(30).cumsum()
fig = plt.figure(figsize=(8, 4))
ax = plt.axes()

ax.set_xticks([0,10,20,30])
ax.set_xticklabels(['cero', 'diez', 'veinte', 'treinta'],rotation=30, fontsize='small')

ax.grid(color='b')
ax.plot(random1, 'k--', marker='o')
ax.plot(random1, color='k', linestyle='dashed', marker='o') # igual a la sentencia anterior
plt.show()
```

### Histogramas
---
Los *histogramas* muestran la frecuencia de cada valor de una serie de números. Usamos el método `hist`.

<a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html" target="blank">https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html</a>

Como vimos anteriormente, existen varios tipos de gráficos de barras.

Veamos como realizar diagramas solo de barras, con barras apiladas y con dos o más variables.

<a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html" target="blank">https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html</a>

### Diagrama de Barras
---

Es el diagrama más simple, donde se gráfica una sola serie usando el método `bar`.

El *primer* parámetro indica la posicion de cada valor de la serie en el eje x, el *segundo* la serie, el *tercero* el ancho de la barra, y se pueden agregar más parámetros.

```python
import numpy as np
import matplotlib.pyplot as plt

womenMeans = (25, 32, 34, 70, 25)

ind = [0,1,2,3,4]#np.arange(5) # crea una lista con numeros entre 0 y 4, las posiciones en x.
width = 1 # ancho de las barras

fig = plt.figure(figsize=(6,2)); ax = plt.axes()

ax.bar(ind, womenMeans, width, color='r') # barra roja. 

ax.set_ylabel('Scores'); ax.set_title('Scores by group')
ax.set_xticks(ind); 
ax.set_xticklabels(['G1', 'G2', 'G3', 'G4', 'G5'])
plt.show()
```

### Diagrama de Barras apiladas
---
Este diagrama representa varios grupos de datos, uno sobre el otro, apilados. En el método `bar` se debe usar el parámetro `bottom` para indicar la barra sobre la cual se apila. 

En el ejemplo, primero creamos la barra roja, y luego la barra azul indicando que debe ir sobre la primera.

```python
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(5)
width = 0.35

fig = plt.figure(figsize=(6,2)); ax = plt.axes()
ax.bar(ind, menMeans, width, color='r') # barra roja. 
ax.bar(ind, womenMeans, width, color='b', bottom=menMeans) # barra azul 

ax.set_ylabel('Scores'); ax.set_title('Scores by group and gender')
ax.set_xticks(ind); ax.set_xticklabels(['G1', 'G2', 'G3', 'G4', 'G5'])
ax.legend(labels=['Men', 'Women'])
plt.show()
```

### Diagrama de Barras varias variables
---

En este caso, las barras de las variables se agregan una al lado de la otra.

Se debe tener en cuenta que la posición de cada barra en el eje x, indicada en el primer parámetro del método `bar`, se debe incrementar para cada barra, así no se superponen. Es recomendable que sea el mismo que el ancho de las barras.

```python
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(5) # crea la lista con numeros entre 0 y 4
width = 0.25 # ancho de las barras

fig = plt.figure(figsize=(6,2)); ax = plt.axes()
ax.bar(ind - width/2, menMeans, width, color='r') # barra roja. 
ax.bar(ind + width/2, womenMeans, width, color='b') # barra azul 

ax.set_ylabel('Scores'); ax.set_title('Scores by group and gender')
ax.set_xticks(ind); ax.set_xticklabels(['G1', 'G2', 'G3', 'G4', 'G5'])
ax.legend(labels=['Men', 'Women'])
plt.show()
```

### Introducción

Los mapas son un tipo de gráfico muy útil para representar los datos geoespaciales.

En general disponemos de dos capas:

### Geopandas - Contextily

---

Como ya vimos, *GeoPandas* es una extensión de Pandas, lo que permite trabajar con *Geoseries* y *GeoDataFrame*, estructuras similares a las *Series* y *Dataframe*, pero con al menos una columna con tipos de datos geoespaciales. Esta columna se la llama por default *geometry*.

Geopandas también extiende los métodos de `MatpLotlib` para graficar, y mediante la libreria `contextily` ofrece mapas donde se grafican los datos.

En este ejemplo, vamos a graficar los pozos petroleros de Argentina sobre el mapa del pais.

Dataset: <a href="http://datos.minem.gob.ar/dataset/produccion-de-petroleo-y-gas-por-pozo" target="blank">http://datos.minem.gob.ar/dataset/produccion-de-petroleo-y-gas-por-pozo</a>

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import descartes
import contextily as ctx
```

```python
df_pozos = pd.read_csv("../Data/capitulo-iv-pozos.csv")
```

```python
df_pozos.head(2)
```

Transformamos las coordenadas numéricas en la forma geométrica *Punto*.

```python
geo_pozos = gpd.GeoDataFrame(df_pozos
                     , geometry=gpd.points_from_xy(df_pozos.coordenadax, df_pozos.coordenaday))
```

```python
geo_pozos.head(2)
```

Ya podemos graficar los pozos. Observemos que se ubican mediante las coordenadas latitud y longitud.

Se usan los métodos que se extienden desde Matplotlib para los títulos, labels, grid y plot.

```python
# Muestra los pozos en un gráfico con latitud y longitud
fig, ax = plt.subplots(figsize=(4,6))
ax.set_title('Pozos')
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
plt.grid()

geo_pozos.plot(ax=ax,
               legend = True,
               alpha=0.3)
plt.show()
```

Ahora sobre el gráfico vamos a agregar la capa del mapa. 

Tenemos que realizar tres pasos adicionales:

- en #1 informamos qué sistema de coordenadas (*CRS*) estamos usando. EPSG:4326 indica que es el sistema que mide latitud y longitud.

- en #2 cambiamos al CRS 3857 (Spheric Mercator). Es el que requiere la libreria `contextily`.

- #3 usamos el método `add_basemap` de la libreria contextily para agregar el mapa.

```python
# Muestra los pozos en un mapa real, con medidas segun escala Spheric Mercator
fig, ax = plt.subplots(figsize=(4,6))
ax.set_title('Pozos')
ax.set_xlabel('x - Spheric Mercator')
ax.set_ylabel('y - Spheric Mercator')
plt.grid()

geo_pozos.crs = "EPSG:4326"                        #1
geo_pozos_3857 = geo_pozos.to_crs(epsg=3857)       #2
geo_pozos_3857.plot(ax=ax,
                   legend = True,
                   alpha=0.3)

ctx.add_basemap(ax, source=ctx.providers.Stamen.TonerLite) #3

#NOTA: el warning se genera para el código geo_pozos.to_crs(epsg=3857).
```
