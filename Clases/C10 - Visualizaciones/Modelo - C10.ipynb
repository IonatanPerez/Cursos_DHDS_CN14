```python
# Para hacer los ejemplos vamos a usar la base de datos de arbolado
import pandas as pd

#arboles = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/arbolado-publico-lineal/arbolado-publico-lineal-2017-2018.csv")
arboles = pd.read_csv("arbolado.csv")

arboles.head()
```

```python
# Vamos a reducir el tama√±o y tomar solo las 10 especies mas frencuentes

arbolado = arboles.sample(1000)
especies = arbolado['nombre_cientifico'].value_counts().index[:10].to_list()
arbolado = arbolado[arbolado['nombre_cientifico'].isin(especies)]

```

```python
import matplotlib.pyplot as plt
# hagamos un histograma de las alturas de los arboles visualizando solo la parte entre 0 y 20 mts. Poner titulo y nombre a los ejes

datos = arbolado['altura_arbol'].values

fig = plt.figure(figsize=(10,6))
ax = plt.axes()

ax.hist(datos, color='steelblue')
ax.set_xlim(0, 20)    

fig.suptitle("Distribucion de alturas de los arboles")
ax.set(xlabel = "altura (m)", ylabel = "frecuencia")
```

```python
# Hacer un diagrama de torta mostrando que porcentaje de arboles corresponde a cada especie. Poner titulo.

valores = arbolado['nombre_cientifico'].value_counts()
etiquetas = arbolado['nombre_cientifico'].value_counts().index

fig = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.pie(valores, labels = etiquetas)
plt.show()
```

```python
# Hacer uns catter plot con seaborn de altura vs diametro separando por especie y sin separar
import seaborn as sns

sns.scatterplot(data=arbolado, y="altura_arbol", x="diametro_altura_pecho")
```

```python
sns.scatterplot(data=arbolado, y="altura_arbol", x="diametro_altura_pecho", hue='nombre_cientifico')
```

```python
# Para hacer los ejemplos vamos a usar la base de datos de arbolado
import pandas as pd

#arboles = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/arbolado-publico-lineal/arbolado-publico-lineal-2017-2018.csv")
arboles = pd.read_csv("arbolado.csv")

arboles.head()
```

```python
# Vamos a reducir el tama√±o y tomar solo las 10 especies mas frencuentes

arbolado = arboles.sample(1000)
especies = arbolado['nombre_cientifico'].value_counts().index[:3].to_list()
arbolado = arbolado[arbolado['nombre_cientifico'].isin(especies)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ejemplo rap√¨do de scatter

sns.set_style('whitegrid')
plt.figure(figsize=(8, 6))
sns.scatterplot(data=arbolado, x="diametro_altura_pecho", y="altura_arbol")
plt.xlabel("diametro tronco")
plt.ylabel("altura")
plt.title("Scatter Plot")
```

```python
# Grafico util para variable numerica vs categorica

sns.set_style('darkgrid')
plt.figure(figsize=(8, 6))
sns.boxplot(data=arbolado, x="nombre_cientifico", y="altura_arbol")

plt.xlabel("nombre cientifico")
plt.ylabel("altura")
plt.title("Box Plot")
```

```python
# Si queremos ver la distribucion en fotmato histograma


plt.figure(figsize=(8, 6))
sns.displot(data=arbolado, x="altura_arbol", hue="nombre_cientifico", multiple="stack", stat="probability")
```
