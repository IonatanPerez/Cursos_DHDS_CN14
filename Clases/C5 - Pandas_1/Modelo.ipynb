## Pandas

Es una librería que agrega funcionalidades sobre Numpy. Principalmente combina la eficiencia en los cálculos numérico y el manejo de arrays con el manejo de diversos tipos de datos que ayudan a hacer la limpieza de datos.

Además incluye un manejo más intuitivo y con etiquetas de la selección de datos.

## Formas de filtrar datos especificas de pandas: Index

Pandas permite elegir los datos para su manipulación de una manera muy intuitiva y mediante etiquetas. Vamos a ver diferentes maneras de usar esas herramientas. Tiene propiedades de conjuntos y es no mutable.

## Un ejemplo con un dataset de arbolado de CABA.

```python
# Importamos librerias

import pandas as pd
```

```python
# Cargamos los datos

#filename = "arbolado-publico-lineal-2017-2018.csv"
data = pd.read_csv('https://cdn.buenosaires.gob.ar/datosabiertos/datasets/arbolado-publico-lineal/arbolado-publico-lineal-2017-2018.csv')

data.head()
```

```python
# Vamos a buscar informacion descriptiva

data.info()
```

```python
data['tipo_activ'].value_counts()
```

```python
data.columns
```

```python
for column in data.columns:
    print (f"Informacion de la columna {column}:")
    display (data[column].describe())
    print ("\n")
```

```python
# Vamos a ver como responder algunas preguntas:

# ¿Donde estan los arboles?

import seaborn as sns

sns.scatterplot(data = data, x="long", y="lat")
```

```python
# Que especies de arboles hay en la ciudad

display (data["nombre_cientifico"].unique())

# Cuantas especies hay?
display (len(data["nombre_cientifico"].unique()))
```

```python
data["nombre_cientifico"].value_counts()
```

```python
# Queremos ver donde estan los arboles "Ficus benjamina"


sns.scatterplot(data = data[data["nombre_cientifico"] == "Melia azedarach"].sample(1000), x="long", y="lat")
```

```python
# Todos los valores son validos?

display (data.tail(20))

```

```python
display (data.isna())
```

```python
display (data.isnull().sum())
```

```python
# Queremos graficar los arboles en algunas comunas.

sns.scatterplot(data = data[data['comuna']==3], x="long", y="lat", hue="comuna")
```

```python
# Vamos a hacer lo mismo pero con menos muestras

sns.scatterplot(data = data.sample(1000), x="long", y="lat", hue="comuna") # Mostrar que pasa cuando se ponen pocas muestras y se ejecuta varias veces
```

```python
data.head()
```

```python
data.index = data["comuna"]
display (data.head())

```

```python
display (data.loc[2])
#display (data.iloc[2])
data = data.rename(columns = {'comuna':'comuna_old'})
data = data.reset_index(drop=False)
display (data.head())
```

```python
# Queremos saber la altura promedio de los arboles de la comuna 3

print (f"La altura promedio de los arboles de la comuna 3 es: {data[data['comuna']==3]['altura_arbol'].mean()}") # Notar que hay un problema si usamos "altura_arbol"

# Lo queremos hacer para cad grupo

for nro_comuna in data['comuna'].unique(): # Ver que pasa si no ponemos unique
    print (f"La altura primedio de los arboles de la comuna {nro_comuna} es: {data[data['comuna']==nro_comuna]['altura_arbol'].mean()}") # Notar que hay un problema si usamos "altura_arbol"
```
