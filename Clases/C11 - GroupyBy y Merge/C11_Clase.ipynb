```python
# Para hacer los ejemplos vamos a usar la base de datos de arbolado
import pandas as pd

#arboles = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/arbolado-publico-lineal/arbolado-publico-lineal-2017-2018.csv")
arboles = pd.read_csv("arbolado.csv")

arboles.head()
```

```python
# Vamos a reducir el tamaño y tomar solo las 10 especies mas frencuentes

arbolado = arboles.sample(1000)
especies = arbolado['nombre_cientifico'].value_counts().index[:3].to_list()
arbolado = arbolado[arbolado['nombre_cientifico'].isin(especies)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# hagamos un scatter entre altura y diametro pecho

# Ejemplo rapìdo de scatter

sns.set_style('whitegrid')
plt.figure(figsize=(8, 6))
sns.scatterplot(data=arbolado, x="diametro_altura_pecho", y="altura_arbol")
plt.xlabel("diametro tronco")
plt.ylabel("altura")
plt.title("Scatter Plot")
```

```python
# Grafico util para variable numerica vs categorica

sns.set_style('darkgrid')
plt.figure(figsize=(8, 6))
sns.boxplot(data=arbolado, x="nombre_cientifico", y="altura_arbol")

plt.xlabel("nombre cientifico")
plt.ylabel("altura")
plt.title("Box Plot")
```

```python
# Si queremos ver la distribucion en fotmato histograma


plt.figure(figsize=(8, 6))
sns.displot(data=arbolado, x="altura_arbol", hue="nombre_cientifico", multiple="stack", stat="probability")
```

# Ejemplos con lo de arbolado

```python
# Para hacer los ejemplos vamos a usar la base de datos de arbolado
import pandas as pd

#arboles = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/arbolado-publico-lineal/arbolado-publico-lineal-2017-2018.csv")
arboles = pd.read_csv("arbolado.csv")

arboles.head()
```

```python
arbolado = arboles

arbolado.shape
```

```python
# Quiero los arboles de los cuales hay una cantidad razonable, osea quiero sacar los arboles muy raros

criterio = 10000

arbolado = arbolado.groupby(['nombre_cientifico']).filter(lambda grp: grp.shape[0] > criterio)

arbolado['nombre_cientifico'].value_counts().tail()
```

```python
# queremos el promedio de altura para cada especie

arbolado.groupby(['nombre_cientifico'])['altura_arbol'].mean()
arbolado.groupby(['nombre_cientifico']).apply(lambda grp: grp['altura_arbol'].mean())
arbolado.groupby(['nombre_cientifico'])['altura_arbol'].aggregate('mean')
arbolado.groupby(['nombre_cientifico']).aggregate({'altura_arbol':'mean'})
```

```python
# Hay datos faltantes en altura?
arbolado['altura_arbol'].isna().sum()

# opcion 1

arbolado_m = arbolado
arbolado_m.index = arbolado_m['nombre_cientifico']
arbolado_m = arbolado_m.drop(columns=['nombre_cientifico']).fillna(arbolado.groupby(['nombre_cientifico'])['altura_arbol'].mean())

```

```python

arbolado_agrupado = arbolado.groupby(['nombre_cientifico','comuna'])
arbolado_altura_para_imputar = arbolado_agrupado['altura_arbol'].transform(lambda x: x.mean())
```
